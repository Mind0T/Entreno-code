```python
import pandas as pd
```

```python
datos=pd.read_csv("cen.csv")
```

```python
datos.info()
```

```python
datos.head()
```

```python
import seaborn as sb
sb.scatterplot(x="celcius",y="kelvin",data=datos,hue="kelvin",palette="coolwarm")
```

```python
x=datos["celcius"]
y=datos["kelvin"]
```

```python
x_procesada=x.values.reshape(-1,1)
y_procesada=y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo=LinearRegression()
```

```python
modelo.fit(x_procesada,y_procesada)
```

```python
modelo.score(x_procesada,y_procesada)
```

```python
cel=int(input("Ingresa una temperatura en celcius y te dire a cuantos kelvin equivale"))
prediccion=modelo.predict([[cel]])
print(f"Ingresaste {cel} grados celcius que equivale a {prediccion} kelvins")
```
