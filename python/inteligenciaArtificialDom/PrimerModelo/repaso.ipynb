```python
import seaborn as sb
sb.scatterplot(x="fahrenheit",y="kelvin",data=df,hue="kelvin", palette="coolwarm")
```

```python
x=df["fahrenheit"]
y=df["kelvin"]
```

```python
x_procesada=x.values.reshape(-1,1)
y_procesada=y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
modelo=LinearRegression()
```

```python
modelo.fit(x_procesada,y_procesada)
```

```python
fahr=float(input("Ingresa una cantidad en fahrenheit y te dire su equivalente en kelvin"))
prediccion=modelo.predict([[fahr]])
print(f"Usted ingreso {fahr} grados fahrenheit que equivalen a {prediccion} kelvins")
```

```python
modelo.score(x_procesada,y_procesada)
```
